## Oh wow, it's a beautiful minestrone soup!

We now have easy access to all kinds of information about the web page through a few user-friendly properties and methods!
The method get_text is useful in only showing the text parts of an entire document or a tag! Once you have text, well... then you can start playing with string functions.

````python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/Soup'
r = requests.get(url)
soup = BeautifulSoup(r.text,'html.parser') #create BeautifulSoup-object from html-text

print(soup.title)                       #get title of page          
print(soup.title.get_text())
print(len(soup.get_text()))             #character count of the text in this page
print(soup.get_text()[0:500].split())   #get all words from the stripped text (only first 500 chars)
print(soup.get_text().count("soup"))    #count all occurences of the word soup

````

## Let's start scraping already!
Now you know how BeautfilSoup handles an HTML page, it's time to start looking for 'interesting' data. We can use the methods find and find_all for this task. __find_all () __ returns all found elements (based on a particular search) while __find() __ returns only the very first found element.

### Soup/Soupe/Soep/Soppa...

I want to find all translations for the word soup, therefore i start looking in the left column where there is a list of all supported languages, so my search begins with inspecting those elements! 

```
<li class="interlanguage-link interwiki-de" >
    <a href="https://de.wikipedia.org/wiki/Suppe" title="Suppe – German" lang="de" hreflang="de" class="interlanguage-link-target">Deutsch</a></li>
```
I will have to scrape the a-tag with class 'interlanguage-link-target'



````python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/Soup'
r = requests.get(url)
soup = BeautifulSoup(r.text) 
a_tags = soup.find_all('a',{ "class" : "interlanguage-link-target" })

for a in a_tags:
    print(a.get('title'))

#or more pythonic ;)
names = [a['title'] for a in a_tags]
print(names)

#with only one split-function you extract the translations of the word soup! 
````

### Get those images

I want to download the interesting images from this page. First open a picture and try to get your head around its url and how it is used in an image-tag.

````python
from bs4 import BeautifulSoup
import requests
import re

url = 'https://en.wikipedia.org/wiki/Soup'
r = requests.get(url)
soup = BeautifulSoup(r.text) 

img_tags = soup.find_all('img')         #all image-tags
src = [img['src'] for img in img_tags]  #only urls

#we use a regex to get the urls to the stored images (without 120px etc...)
#Yes indeed, regex is a huge help in every string situation
result = re.findall(r'(/\w+[.](jpg|gif|png))', str(src)) 

for url in result:
    print(url[0])
    #i did the hard part, so try now to download all the images to your hard drive!
````
